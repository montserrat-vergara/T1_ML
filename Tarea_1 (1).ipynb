### Pregunta 1.14

Detecte las observaciones outliers de las tres variables seleccionadas en la pregunta anterior. Además, impute estas observaciones si usted lo considera necesario. Justifique su decisión.

```python
variables = ['Population', 'Area (sq. mi.)', 'Phones (per 1000)']

# Crear un DataFrame para almacenar los outliers
outliers = pd.DataFrame()

# Calcular el IQR y detectar outliers para cada variable
for var in variables:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    outliers_var = df[(df[var] < limite_inferior) | (df[var] > limite_superior)]
    outliers = pd.concat([outliers, outliers_var])

# Eliminar duplicados que puedan surgir al concatenar
outliers = outliers.drop_duplicates()

print("Outliers detectados:")
print(outliers)
```

```python
# hacer un boxplot para detectar outliers
plt.figure(figsize=(15, 5))

for i, var in enumerate(variables, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df[var])
    plt.title(f'Boxplot de {var}')

plt.tight_layout()
plt.show()
```




---


Se detectaron outliers en las variables Population, Area (sq. mi.) y Phones (per 1000) mediante boxplots y tablas. En los tres casos, los valores extremos corresponden a países grandes, tanto en tamaño como en población o altamente desarrollados, como China, India o Estados Unidos.

Dado que estos valores reflejan condiciones reales y no errores, no se considera necesario imputarlos. Eliminarlos o modificarlos podría reducir la validez del análisis, ya que representan características importantes que explican parte de la varianza del PIB o del desarrollo tecnológico.

Es decir, para el análisis es importante saber que existen diferencias importantes entre países, lo que nos puede resultar en datos outliers, por ejemplo en temas de Área, es claro que países como India, China, Rusia, serán datos outliers, pero esto no quiere decir que la información sea incorrecta.

Para el año 2007, por cada país realice un ranking de las fuentes con más emisiones de CO2 excluyendo las variables `Per Capita` y `Bunker fuels (Not in Total)`. Es decir, asigne un número de 1 a 5 a $\{$ `Solid Fuel`, `Liquid Fuel`, `Gas Fuel`, `Cement`, `Gas Flaring` $\}$, donde 1 es la mayor fuente de emisión de ese país en aquel año, y 5 indica que fue la menor; así para todos los países.

Si en 2007 no se reporta una fuente de emisión para un país, por ejemplo, si emisiones de `Gas Flaring` no se reportara, entonces asigne números de 1 a 4 a las fuentes restantes. Análogo para un menor número de datos.

Luego, por cada variable grafique un histograma de frecuencias del ranking que obtuvo la fuente emisión a lo largo de todos los países.

¿Cuál fue la fuente más contaminante en la mayoría de países en 2007?


```python
# Filtrar datos para el año 2007
df_co2_2007 = df_co2[df_co2['Year'] == '2007']

# Seleccionar columnas relevantes
columnas_emisiones = ['Solid Fuel', 'Liquid Fuel', 'Gas Fuel', 'Cement', 'Gas Flaring']
df_ranking = df_co2_2007[['Country'] + columnas_emisiones]
```

```python
# Calcular el ranking para cada país
def ranking_emisiones(row):
    valores = row[columnas_emisiones].dropna().values #guarda los valores de las emisiones, sin los nulos, en una variable
    columnas_con_datos = row[columnas_emisiones].dropna().index # guarda los nombres de las columnas

    # Chequear si hay datos para las fuentes de emisión, en caso de estar vacias devuelve una serie de "none" values
    if len(valores) == 0:
        return pd.Series([None] * len(columnas_emisiones), index=columnas_emisiones)

    # Crea el ranking de manera ascendente
    ranking = pd.Series(valores, index=columnas_con_datos).rank(ascending=False, method='dense')

    # Aplica el ranking a todas las variables
    ranking = ranking.reindex(columnas_emisiones)

    return ranking

# Aplicamos la función
df_ranking = df_ranking.apply(ranking_emisiones, axis=1, result_type="expand")
```

```python
# Crear histogramas
for columna in columnas_emisiones:
    plt.figure(figsize=(8, 6))
    plt.hist(df_ranking[columna].dropna(), bins=5, range=(1, 5), align="left", rwidth=0.8)
    plt.title(f'Histograma de Ranking para {columna} en 2007')
    plt.xlabel('Ranking')
    plt.ylabel('Frecuencia')
    plt.xticks(range(1, 6))
    plt.show()

# Fuente más contaminante en la mayoría de los países en 2007
ranking_counts = df_ranking[columnas_emisiones].apply(pd.value_counts).fillna(0)
fuente_mas_contaminante = ranking_counts.loc[1].idxmax()
print(f"La fuente más contaminante en la mayoría de los países en 2007 fue: {fuente_mas_contaminante}")
```

Para cada serie de total de emisiones por país, calcule el cambio porcentual a través del tiempo. Realice imputación de missings si considera necesario, justificando su imputación. Si no lo considera necesario, también justifique (se evaluará un buen criterio fundamentado).

```python
# Calcular el cambio porcentual de las emisiones totales de CO2 por país a través del tiempo.
df_co2['Total_pct_change'] = df_co2.groupby('Country')['Total'].pct_change()
df_co2.head()
```

```python
# Justificación de la imputación de valores faltantes:
# Se utilizará la imputación forward fill para los valores faltantes.
# Esto es razonable ya que la serie de tiempo muestra tendencias continuas y la imputación forward fill propaga el último valor observado hacia adelante,
# reduciendo la discontinuidad en la serie y evitando cambios bruscos en la tendencia de los datos.

df_co2['Total_pct_change'] = df_co2.groupby('Country')['Total_pct_change'].ffill()
```

```python
# Filtrar datos para años mayores o iguales a 1995
df_co2_1995 = df_co2[df_co2['Year'].dt.year >= 1995]

# Seleccionar países para graficar
paises = ['United Kingdom', 'Canada', 'Germany', 'France', 'United States', 'Brazil', 'China', 'Japan', 'India']
df_co2_filtrado_1995 = df_co2_1995[df_co2_1995['Country'].isin(paises)]

# Crear el gráfico de cambios porcentuales
plt.figure(figsize=(12, 6))
for pais in paises:
    df_pais = df_co2_filtrado_1995[df_co2_filtrado_1995['Country'] == pais]
    plt.plot(df_pais['Year'], df_pais['Total_pct_change'], label=pais)

plt.xlabel('Año')
plt.ylabel('Cambio Porcentual de Emisiones Totales de CO2')
plt.title('Cambio Porcentual de Emisiones Totales de CO2 por País (1995-2022)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


---


Para calcular el cambio porcentual de emisiones totales de CO₂ por país a lo largo del tiempo, se utilizó una imputación por método forward fill. Esta técnica consiste en rellenar los valores faltantes con el último valor conocido, lo que resulta razonable cuando se espera cierta continuidad temporal en los datos y no hay grandes fluctuaciones año a año. Dado que las emisiones suelen tener una tendencia relativamente estable o creciente, y considerando que muchos de los valores faltantes se encontraban en años cercanos, el uso de forward fill permite mantener la serie continua sin introducir saltos abruptos que podrían distorsionar los cambios porcentuales.
