```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

Cargar la BBDD (CASEN 2022)

```python
# URL base del repositorio (ajusta con tu usuario/repositorio reales)
base_url = "https://raw.githubusercontent.com/montserrat-vergara/T1_ML/main/csv_dividido/parte_{}.csv"

# NÃºmero total de archivos
total_archivos = 41

# Lista para almacenar los DataFrames
dfs = []

for i in range(1, total_archivos + 1):
    url = base_url.format(i)
    print(f"ðŸ“¥ Cargando: parte_{i}.csv")
    dfc = pd.read_csv(url)
    dfs.append(dfc)

# Combinar todos los DataFrames
df = pd.concat(dfs, ignore_index=True)

print(f"\nâœ… Dataset combinado: {len(df)} filas, {df.shape[1]}Â columnas")
```

1.0) Filtramos la BBDD

```python
# Dejar solo personas de 25 aÃ±os o menos
df = df[df["edad"] <= 25]

# Niveles de educaciÃ³n superior a eliminar (ajusta segÃºn codificaciÃ³n real en CASEN)
educacion_sup = [
    "TÃ©cnico nivel superior incompleta",
    "TÃ©cnico nivel superior completo",
    "Profesional incompleto",
    "Profesional completo",
    "Posgrado incompleto",
    "Posgrado completo"
]

df = df[~df["educ"].isin(educacion_sup)]
```

Revisamos los missing values de las variables

```python
# Contar valores nulos por columna
missing_values = df.isnull().sum()

# Calcular el porcentaje de valores nulos
missing_percentage = (missing_values / len(df)) * 100

# Crear un DataFrame con el conteo y el porcentaje
missing_info = pd.DataFrame({'Conteo Nulos': missing_values, '% Nulos': missing_percentage})

# Filtrar para mostrar solo columnas con valores nulos (si quieres)
missing_info = missing_info[missing_info['Conteo Nulos'] > 0]

# Ordenar por el porcentaje de nulos (opcional)
missing_info = missing_info.sort_values(by='% Nulos', ascending=False)

# Imprimir la informaciÃ³n
print("\nInformaciÃ³n de valores nulos por variable:")
print(missing_info)

# Opcionalmente, visualizar los missing values (para un nÃºmero manejable de columnas)
# Si hay muchas columnas con nulos, puede ser difÃ­cil de visualizar.
# Puedes seleccionar un subconjunto o usar una matriz de nulidad.
if missing_info.shape[0] > 0 and missing_info.shape[0] < 50: # LÃ­mite arbitrario para visualizaciÃ³n
    plt.figure(figsize=(12, 6))
    sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
    plt.title('Mapa de Calor de Valores Nulos')
    plt.show()
elif missing_info.shape[0] >= 50:
    print("\nHay muchas columnas con valores nulos para visualizar con heatmap.")
    print("Considere mostrar solo las columnas con el porcentaje mÃ¡s alto de nulos.")
else:
     print("\nNo hay valores nulos en el DataFrame.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Estilo visual
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("Set2")

# ---------- 1. Variables demogrÃ¡ficas y familiares ----------

# Asistencia por sexo
sns.countplot(data=df, x='sexo', hue='e3')
plt.title("Asistencia escolar por sexo")
plt.xlabel("Sexo")
plt.ylabel("Cantidad")
plt.legend(title='Â¿Asiste a clases?', labels=['SÃ­', 'No'])
plt.show()

# Asistencia por zona (rural/urbana)
sns.countplot(data=df, x='area', hue='e3')
plt.title("Asistencia escolar por zona")
plt.xlabel("Zona")
plt.ylabel("Cantidad")
plt.legend(title='Â¿Asiste?', labels=['SÃ­', 'No'])
plt.show()

# Tabla cruzada: vive con padre o madre
print("\nTabla cruzada: asistencia vs. vive con padre/madre (h5_10)")
print(pd.crosstab(df['e3'], df['h5_10'], normalize='columns') * 100)

# Nacionalidad
print("\nTabla: asistencia segÃºn nacionalidad (r1a)")
print(pd.crosstab(df['e3'], df['r1a'], normalize='columns') * 100)

# Pueblos originarios
print("\nTabla: asistencia segÃºn pertenencia a pueblo originario (r3)")
print(pd.crosstab(df['e3'], df['r3'], normalize='columns') * 100)

# OrientaciÃ³n sexual (si se responde)
if 'os1' in df.columns:
    print("\nDistribuciÃ³n de orientaciÃ³n sexual (os1):")
    print(df['os1'].value_counts(dropna=False))


# ---------- 2. SocioeconÃ³mico y zona ----------

# Ingreso (log) segÃºn asistencia
df['log_ingreso'] = df['ytoth'].apply(lambda x: np.log(x + 1))
sns.boxplot(data=df, x='e3', y='log_ingreso')
plt.title("Log ingreso total del hogar segÃºn asistencia escolar")
plt.xlabel("Â¿Asiste a clases?")
plt.ylabel("Log ingreso")
plt.xticks([0,1], ['SÃ­', 'No'])
plt.show()

# NSE
print("\nAsistencia vs. nivel socioeconÃ³mico (nse)")
print(pd.crosstab(df['e3'], df['nse'], normalize='columns') * 100)

# RegiÃ³n
print("\nAsistencia por regiÃ³n:")
print(pd.crosstab(df['e3'], df['region'], normalize='columns') * 100)

# Sistema previsional
print("\nAsistencia segÃºn sistema previsional (s13):")
print(pd.crosstab(df['e3'], df['s13'], normalize='columns') * 100)

# Asistencia y zona
print("\nAsistencia segÃºn zona (rural/urbana):")
print(pd.crosstab(df['e3'], df['area'], normalize='columns') * 100)


# ---------- 3. EducaciÃ³n ----------

# Sabe leer y escribir
print("\nTabla: asistencia segÃºn si sabe leer y escribir (e1):")
print(pd.crosstab(df['e3'], df['e1'], normalize='columns') * 100)

# Nivel educativo mÃ¡s alto al que asistiÃ³
print("\nAsistencia vs. nivel educativo mÃ¡s alto (e6a):")
print(pd.crosstab(df['e3'], df['e6a'], normalize='columns') * 100)

# Dependencia del establecimiento
print("\nAsistencia vs. dependencia del establecimiento (e10):")
print(pd.crosstab(df['e3'], df['e10'], normalize='columns') * 100)

# Rezago escolar
print("\nAsistencia vs. rezago escolar:")
print(pd.crosstab(df['e3'], df['rezago'], normalize='columns') * 100)

# Becas y beneficios
for col in ['e13a']:  # puedes agregar aquÃ­ e12*, e14* si estÃ¡n como columnas separadas
    print(f"\nAsistencia vs. {col} (recibiÃ³ beneficio escolar):")
    print(pd.crosstab(df['e3'], df[col], normalize='columns') * 100)


# ---------- 4. Salud, apoyo, maternidad ----------

# Estado nutricional
print("\nAsistencia vs. estado nutricional (s2):")
print(pd.crosstab(df['e3'], df['s2'], normalize='columns') * 100)

# RecibiÃ³ alimentos pÃºblicos
print("\nAsistencia vs. recibiÃ³ alimentos pÃºblicos (s4):")
print(pd.crosstab(df['e3'], df['s4'], normalize='columns') * 100)

# Edad al tener primer hijo
if 's6' in df.columns:
    sns.boxplot(data=df[df['s6'].notnull()], x='e3', y='s6')
    plt.title("Edad al tener primer hijo segÃºn asistencia")
    plt.xlabel("Â¿Asiste?")
    plt.ylabel("Edad al primer hijo")
    plt.xticks([0,1], ['SÃ­', 'No'])
    plt.show()

```
